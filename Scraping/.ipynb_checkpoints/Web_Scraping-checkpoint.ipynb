### Criando Uma Rotina de Scraping

```python
# importando bibliotecas 
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
import pandas as pd

# declarando variável cards
cads = []

# obtendo o HTML
response = urlopen('https://www.amazon.com.br/s?k=iphone&i=electronics&__mk_pt_BR=%C3%85M%C3%85%C5%BD%C3%95%C3%91&ref=nb_sb_noss_2')
html = response.read().decode('utf-8')
soup = BeautifulSoup(html, 'html.parser')

# obtendo as TAGs de interesse
anuncios = soup.find('div', {'class': 's-main-slot s-result-list s-search-results sg-row'}).findAll('div', class_='sg-col-4-of-24 sg-col-4-of-12 sg-col-4-of-36 s-result-item s-asin sg-col-4-of-28 sg-col-4-of-16 sg-col sg-col-4-of-20 sg-col-4-of-32')

# coletando as informações dos CARDs
for anuncio in anuncios:
    if anuncio.find('span', {'aria-hidden': 'true'}):
        card = {}

        # nomes
        card['name'] = anuncio.find('span', {'class': 'a-size-base-plus a-color-base a-text-normal'}).getText()

        # valor
        card['value'] = anuncio.find('span', {'aria-hidden': 'true'}).getText()

        # adicionando resultado a lista cards
        cards.append(card)

# criando um DataFrame com os resultados
dataset = pd.DataFrame(cards)
dataset.to_csv('Data/DataSet_Amazon.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```
